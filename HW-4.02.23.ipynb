```python
n = int(input())
print("Количество учащихся:" ,n)
s = 0
for i in range(1, n + 1):
    height = int(input())
    print("Рост:", height)
    s += height
print("Средний рост:" ,s / n)
```

```python
t = int(input())
print(t, "минут")
s = 1
for i in range(t):
    s *= 2
print("Количество бактерий:", s)
```
